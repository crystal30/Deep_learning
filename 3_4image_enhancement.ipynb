```python
# resize
# crop
# flip
# brightness & contrast

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
img_name = './palace.jpeg'
img_string = tf.io.read_file(img_name)
img_decode = tf.image.decode_image(img_string)

%matplotlib inline
print(img_decode.shape)
plt.imshow(img_decode)
```

```python
# resize

# tf.image.resize_area
# tf.image.resize_bicubic
# tf.image.resize_nearest_neighbor

img_name = './palace.jpeg'
img_string = tf.io.read_file(img_name)
img_decode = tf.image.decode_image(img_string)

resize_img = tf.image.resize(img_decode, [250, 576])
resize_img = np.asarray(resize_img, np.uint8)
%matplotlib inline
print(resize_img.shape)
plt.imshow(resize_img)
```

```python
# crop

# tf.image.pad_to_bounding_box
# tf.image.crop_to_bounding_box
# tf.random_crop

img_name = './palace.jpeg'
img_string = tf.io.read_file(img_name)
img_decode = tf.image.decode_image(img_string)

padded_img = tf.image.pad_to_bounding_box(img_decode, 50, 100, 500, 800)
padded_img = np.asarray(padded_img, np.uint8)
%matplotlib inline
print(padded_img.shape)
plt.imshow(padded_img)
```

```python
# flip
# tf.image.flip_up_down
# tf.image.flip_left_right
# tf.image.random_flip_up_down
# tf.image.random_flip_left_right

img_name = './palace.jpeg'
img_string = tf.io.read_file(img_name)
img_decode = tf.image.decode_image(img_string)

flipped_img = tf.image.flip_up_down(img_decode)
flipped_img = np.asarray(flipped_img, np.uint8)
%matplotlib inline
print(flipped_img.shape)
plt.imshow(flipped_img)
```

```python
# brightness
# tf.image.adjust_brightness
# tf.image.random_brightness
# tf.image.adjust_constrast
# tf.image.random_constrast

img_name = './palace.jpeg'
img_string = tf.io.read_file(img_name)
img_decode = tf.image.decode_image(img_string)

brightness_img = tf.image.adjust_brightness(img_decode, 0.5)
brightness_img = np.asarray(brightness_img, np.uint8)
%matplotlib inline
print(brightness_img.shape)
plt.imshow(brightness_img)
```
